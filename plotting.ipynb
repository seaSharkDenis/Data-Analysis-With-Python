```python
import matplotlib.pyplot as plt

# Sample data for bar chart
counties = ["Nyair", "Neighrobi", "Matchekos", "Moanasa", "Karaniaga"]
population = [20000,23450,25000,26250,29775]

# Sample data for line chart
names = ["Kennedy", "Abel", "Lilian", "Cassie", "Jane"]
age = [21,43,29,26,40]

# Sample data for histogram
ages = [22, 21, 20, 23, 24, 22, 20, 21, 22, 25, 23]
```

```python
# Simple line plot
# age vs. name
plt.plot(names,age)
plt.show()
```

```python
# Bar Plot
# counties v. population
plt.bar(counties,population)
plt.show()
```

```python
# Histogram
# counties v. population
plt.hist(ages, bins=5, color='purple')
plt.show()
```
